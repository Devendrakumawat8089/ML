```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("cars.csv")
```

```python
df.head(2)
```

```python
df.drop(columns= ['name'] , inplace=True)
```

```python
df['owner'] = df['owner'].map({"First Owner":0 , "Second Owner":1 , "Third Owner":2 , "Fourth & Above Owner":3 , "Test Drive Car":4 })
df['transmission'] = df['transmission'].map({"Manual":0 , "Automatic":1 })
df['seller_type'] = df['seller_type'].map({"Individual":0 , "Dealer":1 ,"Trustmark Dealer":2})
df['fuel'] = df['fuel'].map({"fuel":0 , "Diesel":1 ,"Petrol":2 , "CNG":3 , "LPG":4 , "Electric":5})
```

```python
df
```

```python
x = df.drop(columns='selling_price')
y = df['selling_price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y ,test_size=0.2 , random_state=42)
```

```python
np.round(x_train.describe(),2)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mn = MinMaxScaler()
```

```python
x_train_mn = mn.fit_transform(x_train)
```

```python
x_train_new = pd.DataFrame(x_train_mn, columns= x_train.columns)
```

```python
np.round(x_train_new.describe(),2)
```
