```python
# Ordinal Encoding --->
# Label Encoding ---> column[sub1,syub2,sub3...subn] ---> encode
# Ordinal Encoding ---> column[sub1,sub2,...subn] ---> Re-Arrange ---> encode
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("covid_toy.csv")
# df.head(2)
```

```python
df = df.dropna()
```

```python
df = df.drop(columns = ['city', 'age', 'fever'])
```

```python
# df.head(2)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Male','Female'],['Mild','Strong'],['No','Yes']])
```

```python
oe_ar = oe.fit_transform(df)
```

```python
new_df = pd.DataFrame(oe_ar, columns = df.columns)
```

```python
# LabelEncoder ---> when we have single column . we cannot set sub-category form .
# has_covid['yes','no'] ---> labelecoder --> [1,0]
# OrdinalEncoder ---> when we have multiple columns .we can set sub-category form.
# has_covid['yes','no'] ---> ord. encoder ---> [0,1]
```

```python
# first approach ::::
# df ---> categorical column ---> ordinal encoder ---> data transform
# second approach ::::
# df ---> x,y --->
# x ---> encode by Ordinal Encoder
# y ---> encode by Label Encoder
```
